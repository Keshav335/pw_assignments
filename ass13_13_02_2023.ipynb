```python
#Q2 Write a python program to print python Exception Hierarchy
print("\nException hierarchy:")
print("---------------------")
for cls in Exception.__subclasses__():
    print(cls.__name__)
    for subcls in cls.__subclasses__():
        print("  {}".format(subcls.__name__))
```

```python
#Q4 Why LookupError class is used? Explain with an example KeyError and IndexError.
#The LookupError class in Python is a base class for all lookup-related errors. It is used to represent the situations where an index or 
#a key is not found in a container like a list or a dictionary.The KeyError: is a type of LookupError that is raised when a dictionary key 
#is not found.
nameRoll = {'Jon': 1, 'Jony': 2, 'Rony': 3}
print(nameRoll['Jack'])
```

```python
#The IndexError is type of LookupError that is raised when an index is out of range.
myList = [1, 2, 3]
print(myList[3])
```
